```python
import pandas as pd
import csv
import re
from io import StringIO
import os
```

```python

reddit = pd.read_csv('reddit_comments_fixed.csv')
reddit.rename(columns={'created_utc': 'date'}, inplace=True)
reddit.rename(columns={'comment_id': 'id'}, inplace=True)
reddit.rename(columns={'score': 'likes'}, inplace=True)
reddit.rename(columns={'body': 'text'}, inplace=True)
reddit['date'] = pd.to_datetime(reddit['date'])
reddit['date'] = reddit['date'].dt.date
# reddit = reddit.drop(columns=['author'])
# reddit = reddit.drop(columns=['post_id'])
reddit = reddit[['id', 'likes', 'date', 'text']]
reddit = reddit[reddit['likes'] >= 2]
reddit.head()

# len(reddit)
# reddit.to_csv('reddit_comments_fixed.csv', index=False)
```

```python
redditPosts = pd.read_csv('reddit_posts_fixed.csv')
# "post_id","title","author","score","created_utc","num_comments","selftext","url"
redditPosts.rename(columns={'post_id': 'id', 'score': 'likes', 'created_utc': 'date', 'selftext': 'text'}, inplace=True)
# redditPosts = redditPosts.drop(columns=['title'])
# redditPosts = redditPosts.drop(columns=['author'])
# redditPosts = redditPosts.drop(columns=['num_comments'])
# redditPosts = redditPosts.drop(columns=['url'])
redditPosts = redditPosts[['id', 'likes', 'date', 'text']]
redditPosts['date'] = pd.to_datetime(redditPosts['date'])
redditPosts['date'] = redditPosts['date'].dt.date
redditPosts = redditPosts[redditPosts['likes'] >= 2]
redditPosts = redditPosts[redditPosts['text'].str.len() >= 7]
redditPosts.head()
# redditPosts.to_csv('reddit_posts_fixed.csv', index=False)

```

```python
youtube = pd.read_csv('youtube_comments_fixed.csv')
#"author","published_at","updated_at","like_count","text"
# youtube.drop(columns=['updated_at'], inplace=True)
youtube.rename(columns={'author': 'id'}, inplace=True)
youtube.rename(columns={'published_at': 'date'}, inplace=True)
youtube.rename(columns={'like_count': 'likes'}, inplace=True)
youtube['date'] = pd.to_datetime(youtube['date'])
youtube['date'] = youtube['date'].dt.date
youtube = youtube[['id', 'likes', 'date', 'text']]
# youtube = youtube[youtube['likes'] >= 2]


keywords = ['grok', 'elon', 'musk', 'xai' , 'ai']
youtube = youtube[youtube['text'].str.contains('|'.join(keywords), case=False, na=False)]
youtube.head()
len(youtube)
# youtube.to_csv('youtube_comments_fixed.csv', index=False)
```
